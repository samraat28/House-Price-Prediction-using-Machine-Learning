```python
# Write a function enabled python code to calculate the final price after applying
# discount. 


def discount_calculator(price,discount):
    discount_amount = (price * discount)/100
    final_price = price - discount_amount
    return final_price
    
a = float(input("Enter the purchase amount:"))
b = int(input("Enter the discount %:"))

result = discount_calculator(a,b)
print("The final price to pay is:",result)
```

```python
# WPC with functions that is going to count the words starting with Capital letters.

def count_capital(sentence):
    words = sentence.split()
    count=0

    for i in words:
        if i[0].isupper():
            count = count+1

    return count


a= input("Enter your Sentence:")

result = count_capital(a)
print(result)
    
```
